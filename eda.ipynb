Starting with the visualisations part

The dataset shows imbalance, with approximately 64% placed and 36% not placed.

```python
sns.countplot(x='Placement_Status',data=df)
plt.title("Placement status distribution")
plt.show()
```

From the CGPA distribution histogram, we observe that most students have CGPA between 6.0 and 7.6. The frequency gradually decreases as CGPA increases beyond 8.0. This indicates that the majority of students fall in the mid-performance category, with fewer students achieving very high CGPA values.

```python
sns.histplot(df['CGPA'],kde=True)
plt.title("CGPA")
plt.show()
```

The correlation heatmap shows a strong positive relationship among academic performance indicators such as CGPA, Aptitude Score, Coding Score, and HR Score (correlation > 0.95). This suggests multicollinearity among features. Placement Status shows strong positive correlation with Coding Score (0.87), Aptitude Score (0.86), and Attention Span (0.86), indicating that technical and cognitive skills significantly influence placement outcomes.

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

The Year vs Placement analysis shows that placement distribution is nearly similar for 3rd and 4th year students. This suggests that academic year alone does not significantly influence placement outcomes in the dataset.

```python
sns.countplot(x='Year', hue='Placement_Status', data=df)
plt.title("Year vs Placement")
plt.show()
```

```python
sns.boxplot(x='Placement_Status', y='Study_Hours_Coding', data=df)
plt.title("Coding Study Hours vs Placement")
plt.show()
```
